```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# importing file of graphs without clique
with open('cliqueless graphs/currentComparisonObject_cliqueLessGraphs.json') as current_comparison_cliqueless_file:
    current_comparison_cliqueless = json.load(current_comparison_cliqueless_file)

# importing file of couple of graphs (with / without clique corrected )
with open('couple of graphs/currentComparisonObject_coupleOfGraphs.json') as current_comparison_couple_file:
    current_comparison_couple = json.load(current_comparison_couple_file)

# setting figure size:
fig, ax = plt.subplots(figsize=(6,4))

# Extract the data that has to be plotted
x = current_comparison_couple["arrayOfCliqueSizes"]
y_noclique_uncorrected = current_comparison_cliqueless["uncorrectedGraphsAverageDegreeArray"]
y_noclique_corrected = current_comparison_cliqueless["correctedGraphsAverageDegreeArray"]
y_clique = current_comparison_couple["graphsWithCliqueAverageDegreeArray"]

# create dataframe with x, y_corrected and y_uncorrected as columns:
df = pd.DataFrame({'x': x, 'y_noclique_corrected': y_noclique_corrected, 'y_noclique_uncorrected': y_noclique_uncorrected, 'y_clique': y_clique})

# Plot the data
ax = sns.lineplot(data=df, x="x", y="y_clique", label="With clique", color='blue')
ax = sns.lineplot(data=df, x="x", y="y_noclique_corrected", label="No clique corrected", color='red', linestyle='dashed')
ax = sns.lineplot(data=df, x="x", y="y_noclique_uncorrected", label="No clique uncorrected", color = 'red')
ax.set_ylabel("Average Degree")
ax.set_xlabel("Clique Size")
ax.legend()
#fig.show()

# Save plot as high resolution png:
fig.savefig('N300_comparisonGraph.png', dpi=300)

# WITH CLIQUE (blue) - NO CLIQUE CORRECTED (red dashed) - NO CLIQUE UNCORRECTED (red filled) 
```
