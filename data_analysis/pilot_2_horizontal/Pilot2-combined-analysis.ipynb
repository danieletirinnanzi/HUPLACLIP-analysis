1. MACHINES:

```python
# CONSTRUCTING UNIQUE DATAFRAME FOR MACHINES DATA:
df_machines_appended = []
N_range = [300,400,500,600,700,800,900,1000]
for N in N_range:
    K = eval(f'x_{300}')*N
    df_machines_appended.append(pd.DataFrame({'N':np.full(K.size, N), 'K':K, 'fc':np.array(eval(f'y_{N}'))/100}))
    
# creating single df for machines:    
df_machines = pd.concat(df_machines_appended)
# creating new variables and adding them to the dataframe:
df_machines['K/sqrt(N)'] = df_machines['K']/np.sqrt(df_machines['N'])
# dropping the indices so that there are no duplicates:
df_machines = df_machines.reset_index(drop=True)
df_machines
```

```python
# FITTING THE MACHINES DATA:
# defining the sigmoid function to fit to the data:
def sigmoid(k, k0, delta):
    return (1+ndtr((k-k0)/delta))/2
# fitting the data to the sigmoid function:
fitted_params_machines = []
for N in N_range:
    data_for_fit = df_machines[df_machines['N']==N]
    fitted_params_machines.append(curve_fit(sigmoid, data_for_fit['K'], data_for_fit['fc'], p0=(70, 20))[0])
fitted_params_machines = np.array(fitted_params_machines)
fitted_params_machines
```

```python
# Creating dataframe of fitted parameters:
df_machines_fitted_params = pd.DataFrame(columns=['N', 'K0', 'Delta'])
dfs = []
# Loop through the array and populate the dataframe
for i in range(len(N_range)):
    N = N_range[i]
    K0_value = fitted_params_machines[i, 0]
    delta_value = fitted_params_machines[i, 1]
    df = pd.DataFrame({'N': [N], 'K0': [K0_value], 'Delta': [delta_value]})
    dfs.append(df)
df_machines_fitted_params = pd.concat(dfs, ignore_index=True)
df_machines_fitted_params
```

```python
# fitting linearly K0 as a function of N and storing slope and intercept:
slope, intercept, r_value, p_value, std_err = linregress(df_machines_fitted_params['N'], df_machines_fitted_params['K0'])
```

```python
# MACHINES PSYCHOMETRICS GRAPH
# Create a larger figure with specified size
fig, ax = plt.subplots(figsize=(6, 5))

# defining x values
x_vals = np.linspace(0, 500, 1000)

# creating palette for the graph
my_palette_machines = sns.color_palette("crest", fitted_params_machines.shape[0])

# Scatter plot
sns.scatterplot(data=df_machines, x='K', y='fc', hue='N', palette=my_palette_machines, ax=ax)

# defining y values for each N value and plotting it
for index in range(fitted_params_machines.shape[0]):
    y_vals = (1 + ndtr((x_vals - fitted_params_machines[index][0]) / fitted_params_machines[index][1])) / 2
    ax.plot(x_vals, y_vals, color=my_palette_machines[index])

# setting labels and title
ax.set_xlim(ax.get_xlim()[0], 510)
ax.set_ylim(0.4, 1.03)
ax.set_ylabel("Fraction correct")
ax.set_xlabel("Clique size")    
ax.tick_params(axis='x')
ax.tick_params(axis='y')

# Remove y ticks that have no meaning
yticks = ax.get_yticks()
yticks = [tick for tick in yticks if tick <= 1]
ax.set_yticks(yticks)

# Change legend name to "number of nodes"
handles, labels = ax.get_legend_handles_labels()
legend = ax.legend(handles, labels, title="Number of nodes", loc='lower right')

# Drawing K0 and delta for N=1000
line_x = fitted_params_machines[7][0]
line_color = my_palette_machines[1]
y_vals = (1 + ndtr((x_vals - fitted_params_machines[7][0]) / fitted_params_machines[7][1])) / 2
line_y = np.interp(line_x, x_vals, y_vals)

# Draw vertical dashed line that stops at the curve
ax.plot([line_x, line_x], [0, line_y], color=my_palette_machines[7], linestyle='--', label='K0')

# Add label "K0" next to the vertical line
ax.text(130, 0.5, 'K', color=my_palette_machines[7], ha='left', va='top')

# Add horizontal error bar
delta_bar_width = fitted_params_machines[7][1]
ax.errorbar(line_x, line_y, xerr=delta_bar_width/2, color=my_palette_machines[7], capsize=3)

# Add label "delta" next to the vertical line
ax.text(160, 0.74, r'$\Delta$', color=my_palette_machines[7], ha='left', va='bottom')

# Decrease scatter plot marker size
scatter_collection = ax.collections[0]
scatter_collection.set_sizes([15])

# saving figure in folder "figures" with 300dpi
fig.savefig('./figures/machines_psychometrics.png', dpi=300)

# Show the plot
plt.show()
```
