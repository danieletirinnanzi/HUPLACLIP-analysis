# 1. MACHINES:

```python
# CONSTRUCTING UNIQUE DATAFRAME FOR MACHINES DATA:
df_machines_appended = []
N_range = [300,400,500,600,700,800,900,1000]
for N in N_range:
    K = eval(f'x_{300}')*N
    df_machines_appended.append(pd.DataFrame({'N':np.full(K.size, N), 'K':K, 'fc':np.array(eval(f'y_{N}'))/100}))
    
# creating single df for machines:    
df_machines = pd.concat(df_machines_appended)
# creating new variables and adding them to the dataframe:
df_machines['K/sqrt(N)'] = df_machines['K']/np.sqrt(df_machines['N'])
# dropping the indices so that there are no duplicates:
df_machines = df_machines.reset_index(drop=True)
df_machines
```

Fitting data for machines:

```python
# defining the sigmoid function to fit to the data:
def sigmoid(k, k0, delta):
    return (1+ndtr((k-k0)/delta))/2
# fitting the data to the sigmoid function:
fitted_params_machines = []
for N in N_range:
    data_for_fit = df_machines[df_machines['N']==N]
    fitted_params_machines.append(curve_fit(sigmoid, data_for_fit['K'], data_for_fit['fc'], p0=(70, 20))[0])
fitted_params_machines = np.array(fitted_params_machines)
fitted_params_machines
```

```python
# Creating dataframe of fitted parameters:
df_machines_fitted_params = pd.DataFrame(columns=['N', 'K0', 'Delta'])
dfs = []
# Loop through the array and populate the dataframe
for i in range(len(N_range)):
    N = N_range[i]
    K0_value = fitted_params_machines[i, 0]
    delta_value = fitted_params_machines[i, 1]
    df = pd.DataFrame({'N': [N], 'K0': [K0_value], 'Delta': [delta_value]})
    dfs.append(df)
df_machines_fitted_params = pd.concat(dfs, ignore_index=True)
df_machines_fitted_params
```

```python
# fitting linearly K0 as a function of N and storing slope and intercept:
slope, intercept, r_value, p_value, std_err = linregress(df_machines_fitted_params['N'], df_machines_fitted_params['K0'])
```

```python
# FITTING THE HUMAN DATA subject by subject:
# defining the sigmoid function to fit to the data:
def sigmoid(k, k0, delta):
    return (1+ndtr((k-k0)/delta))/2

# empty dataframe where fitted parameters will be stored:
df_humans_fitted_params_appended = []
# looping through all subjects:
for subject in df_humans_subjectpooling['subject_number'].unique():
    # looping through N values (in this case, only 300 and 1000):
    for N_value in df_humans_subjectpooling['N'].unique():
        # accessing data for current subject and N value:
        current_data = df_humans_subjectpooling[(df_humans_subjectpooling['subject_number']==subject) & (df_humans_subjectpooling['N']==N_value)]
        # calculating fitted parameters:
        fitted_params = curve_fit(sigmoid, current_data['K'], current_data['fc'], p0=(70, 20))[0]
        # adding fitted parameters to the dataframe:
        df_humans_fitted_params_appended.append(pd.DataFrame({'subject_number':[subject], 'N': [N_value], 'K0':[fitted_params[0]], 'Delta':[fitted_params[1]]}))

# concatenating dataframe of fitted parameters:
df_humans_fitted_params_subjectpool = pd.concat(df_humans_fitted_params_appended)
df_humans_fitted_params_subjectpool
```

```python
# function to find the K value that is closer to K0:
def find_closest_value(arr, target):
    closest_value = arr[0]  # Initialize with the first element
    min_difference = abs(target - arr[0])  # Initialize with the difference to the first element
    
    # every time a closer element is found, the closest value is updated
    for element in arr:
        difference = abs(target - element)
        if difference < min_difference:
            min_difference = difference
            closest_value = element
    
    return closest_value

# find clique values closest to K0 for N=300 and N=1000:
K_300_closest = find_closest_value(df_humans_300_response_times['clique_size'].unique(), K0_300_mean_subjpool)
K_1000_closest = find_closest_value(df_humans_1000_response_times['clique_size'].unique(), K0_1000_mean_subjpool)

# find smallest values of clique size:
K_300_min = df_humans_300_response_times['clique_size'].min()
K_1000_min = df_humans_1000_response_times['clique_size'].min()

# check that response times at K_closest are larger than response times at K_min with a paired sample t-test (one tailed, the alternative hp is that the response times at K_closest are larger than the response times at K_min):
result_300_rt = ttest_rel(df_humans_300_response_times[df_humans_300_response_times['clique_size'] == K_300_closest]['response_time_log_avg_msec'],df_humans_300_response_times[df_humans_300_response_times['clique_size'] == K_300_min]['response_time_log_avg_msec'], alternative='greater')
result_1000_rt = ttest_rel(df_humans_1000_response_times[df_humans_1000_response_times['clique_size'] == K_1000_closest]['response_time_log_avg_msec'],df_humans_1000_response_times[df_humans_1000_response_times['clique_size'] == K_1000_min]['response_time_log_avg_msec'], alternative='greater')

# printing results:
print("N=300")
print(result_300_rt.statistic)
print(result_300_rt.pvalue)
print("N=1000")
print(result_1000_rt.statistic)
print(result_1000_rt.pvalue)
```

```python
# accessing csv files for N = 300 and storing subject number:
df_humans_300_appended = []
# accessing all json files in 'data' directory:
files = Path('./N_300/data/').glob('*.csv')
for file_index, file in enumerate(files):
    # reading single csv file as dataframe
    df = pd.read_csv(file)
    
    # cleaning single file to access the final data:
    # - isolating experiment trials ("canvas-keyboard-response" ones) and dropping irrelevant variables:
    df_cleaned = df[df.trial_type == "canvas-keyboard-response"]
    # - dropping irrelevant variables:
    df_cleaned = df_cleaned.drop(["timeout","failed_images","failed_audio","failed_video","view_history","trial_index", "time_elapsed","internal_node_id" ,"success","stimulus","graphs_couple"], axis=1).copy()
    # - isolating trials were final responses were given (shuffles have " " as response):
    df_final = df_cleaned[df_cleaned['response'].isin(['arrowright','arrowleft'])]
    
    # retrieving subject number for current file:
    subject_number = eval(f'{file_index}')
    
    # retrieving data of current clique size and calculating fraction correct for each K value:
    K_range = df_final['clique_size'].unique()
    for K_value in K_range:
        # isolating data of current clique size:
        current_data = df_final[df_final.clique_size == K_value]
        # calculating accuracy for current clique size (accuracy, when present, is either True or False)
        fraction_correct_current_clique_size = sum(current_data.accuracy) / sum(~ current_data.accuracy.isna())
        # adding data to the dataframe:
        df_humans_300_appended.append(pd.DataFrame({'subject_number':[subject_number+1],'N':[300], 'K':[K_value], 'fc':[fraction_correct_current_clique_size]}))
    
# creating single df for humans at N=300:    
df_humans_300 = pd.concat(df_humans_300_appended)
# creating new variable and adding it to the dataframe:
df_humans_300['K/sqrt(N)'] = df_humans_300['K']/np.sqrt(df_humans_300['N'])
# creating new variable and adding it to the dataframe:
df_humans_300['K/N'] = df_humans_300['K']/df_humans_300['N']
# visualizing dataframe:
display(df_humans_300)
```
