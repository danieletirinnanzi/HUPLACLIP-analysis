```python
import matplotlib.pyplot as plt
import numpy as np
from math import log, sqrt, e

# Generate N values from 0 to 3000
N = np.arange(0, 3001)

# Calculate y values for the first formula: 2 * log2(N)
K_it = 2 * np.log2(N)

# Calculate y values for the second formula: sqrt(N / e)
K_comp = np.sqrt(N / e)

# Create the plot
plt.plot(N, K_it, label='$2\\log_2(N)$')
plt.plot(N, K_comp, label='$\\sqrt{\\frac{N}{e}}$')

# Add labels and title to the plot
plt.xlabel('N')
# Set the y-label and the orientation to horizontal
plt.ylabel('K', rotation='horizontal')
plt.title('Statistical and computational limits')

# Add a legend
plt.legend()

# Save plot as svg
plt.savefig('stat_comp_gap.svg', format='svg')

# Show the plot
plt.show()
```
